```python
from pytrends.request import TrendReq
import pandas as pd
import plotly.express as px
import json
```

```python
path = 'D:\data\Gentrification_DA\Work_SG\list.json'
with open(path, 'r', encoding='utf-8') as f:
    json_data = json.load(f)

print(json_data)
```

```python
kw = []
for i in list(json_data.keys()):
    for j in list(json_data[i].keys()):
        kw.append(j)

print(kw)
```

```python
pytrends = TrendReq(hl = 'ko-kr', tz = 540)
kw_ = kw[:5]

pytrends.build_payload(kw_list=kw_, cat=0, timeframe='2004-01-01 2023-01-01', geo='KR')
data = pytrends.interest_over_time()
data = data.reset_index()

df = pd.DataFrame(data)
df = df.drop('isPartial', axis=1)

print(df)
```

```python
fig = px.line(df, x='date', y=kw_)
fig.show()
```
